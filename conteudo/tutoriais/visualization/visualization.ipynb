## Importando as bibliotecas

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt 
```

## Carregando os dados

```python
# DATASET_PATH = 'D:\\MyProjects\\GItHub\\SI\\conteudo\\tutoriais\\visualization'
DATASET_PATH = os.getcwd()
DATASET_NAME = 'housing.csv'
```

```python
def load_data(dataset_path=DATASET_PATH, dataset_name=DATASET_NAME):
    csv_path = os.path.join(dataset_path, dataset_name)
    return pd.read_csv(csv_path)
```

```python
housing = load_data(DATASET_PATH,DATASET_NAME)
```

### Salvando as figuras

```python
IMAGES_PATH = os.path.join(DATASET_PATH, "images")
print(IMAGES_PATH)

def save_fig(fig_name, tight_layout=True, fig_extension="png", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_name + "." + fig_extension)
    print("Saving figure", fig_name)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

```python
housing.hist(bins=50, figsize=(20,15))
save_fig('histogram_plots')
```

## Visualizando os dados

```python
housing.plot(kind='scatter', x='longitude', y='latitude')
save_fig('visualization_plot')
```

```python
housing.plot(kind="scatter", x="longitude", y="latitude", alpha=0.01)
save_fig("visualization_plot_2")
```

```python
housing.plot(kind='scatter', x='longitude', y='latitude', alpha=0.3,
    s=housing['population']/100, label='population', figsize=(10,7),
    c='median_house_value', cmap=plt.get_cmap('jet'), colorbar=True,
    sharex=False)
plt.legend()
save_fig("visualization_plot_3")
```
